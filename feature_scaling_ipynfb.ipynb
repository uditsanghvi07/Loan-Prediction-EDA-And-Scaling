**Normalization min max scaler **

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

# Load the dataset using the correct file path
file_path = "loan (1).csv"
dataset = pd.read_csv(file_path)

# Handle missing values in 'CoapplicantIncome'
# We use .fillna() to replace any missing values with the mean of the column.
if dataset['CoapplicantIncome'].isnull().any():
    dataset['CoapplicantIncome'] = dataset['CoapplicantIncome'].fillna(dataset['CoapplicantIncome'].mean())

# Apply Min-Max Scaling
# Create an instance of the scaler
scaler = MinMaxScaler()
# The .fit_transform() method learns the min and max values from the data and then scales it
dataset['CoapplicantIncome_minmax'] = scaler.fit_transform(dataset[['CoapplicantIncome']])

# Visualize the distributions before and after scaling
plt.figure(figsize=(12, 6))

# Plot for the original, unscaled data
plt.subplot(1, 2, 1)
plt.title("Before Scaling")
sns.histplot(dataset['CoapplicantIncome'], kde=True)

# Plot for the scaled data
plt.subplot(1, 2, 2)
plt.title("After Min-Max Scaling")
sns.histplot(dataset['CoapplicantIncome_minmax'], kde=True)

# Adjust layout and show the plots
plt.tight_layout()
plt.show()

# Display the first few rows to show the new scaled column
print(dataset[['CoapplicantIncome', 'CoapplicantIncome_minmax']].head())
```
